# <center>Data Cleaning</center>

#### Renameing 'Country Code' to'country_id', 'TableName' to 'country_name':

```python
arable_countries.rename(columns={'Country Code': 'country_id', 'TableName': 'country_name'},inplace=True)
#Test code 
#arable_countries
```

#### Createing  the DataFrame for Income :

```python
income_group = pd.DataFrame(arable_countries['IncomeGroup'])

income_group_copy=pd.DataFrame(income_group['IncomeGroup'])


income_group_copy.replace(['High income'],['HI'],inplace=True)
income_group_copy.replace(['Low income'],['LI'],inplace=True)
income_group_copy.replace(['Middle income'],['MI'],inplace=True)
income_group_copy.replace(['Lower middle income'],['LMI'],inplace=True)
income_group_copy.replace(['Upper middle income'],['UMI'],inplace=True)
income_group_copy.rename(columns={'IncomeGroup': 'g_id'},inplace=True)
income_group.rename(columns={'IncomeGroup': 'ig_desc'},inplace=True)



#ig_df =pd.DataFrame(income_group.join(income_group_copy))
ig_df =pd.DataFrame(income_group_copy.join(income_group))

#test code
#ig_df


```

#### Createing  the DataFrame for Countries

```python
countries_info =arable_countries.filter(['country_id','country_name','Region'],axis=1)
countries_df=pd.DataFrame(countries_info.join(ig_df['g_id']))

#Test code
#ountries_info
countries_df                                                   
                                                    
```
